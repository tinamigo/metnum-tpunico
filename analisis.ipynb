```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results.csv')
df =df.sort_values(by='alfa')
```

```python
for k in range(5,12):
    df_k = df[df['k'] == k]
    df_k = df_k[df_k['alfa'] > 15]
    plt.plot(df_k['alfa'],df_k['acc'], '-o', label=k)
plt.legend()
```
